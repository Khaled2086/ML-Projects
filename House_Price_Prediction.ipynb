Importing the required liabriries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Loading the Boston price data set into a dataframe using pandas

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/House price prediction/boston.csv',index_col=False)
df.head()
```

Examining the data

```python
df.info()
```

```python
df.drop(columns='Unnamed: 0',inplace=True, axis=1)
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

Understanding the correlation

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':12},cmap='coolwarm')
```

Spliting the data

```python
x=df.drop(columns='PRICE',axis=1)
y=df['PRICE']

print(x.shape, y.shape)
```

Spliting the data into train and test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
print(x_train.shape,x_test.shape)
```

Model training

```python
model=XGBRegressor()
model.fit(x_train,y_train)
```

```python
prediction_train=model.predict(x_train)
print(prediction_train)
```

```python
# measuing the R squared error [RSE] for trained data
r2_score_train=metrics.r2_score(y_train, prediction_train)
print(r2_score_train)
```

```python
# mean absolute error for trained data
abs_score_train=metrics.mean_absolute_error(y_train, prediction_train)
print(abs_score_train)
```

```python
# visualizing the actual price and predicted price
plt.scatter(y_train,prediction_train)
plt.xlabel('actual price')
plt.ylabel('predicted price')
plt.title('actual price vs predicted price')
plt.show()
```

```python
# prediction for test data
prediction_test=model.predict(x_test)
r2_score_test=metrics.r2_score(y_test, prediction_test)
print(r2_score_test)
abs_score_test=metrics.mean_absolute_error(y_test, prediction_test)
print(abs_score_test)
```
