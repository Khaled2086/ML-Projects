```python
# Import the required libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset to dataframe
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Diabetic/diabetes.csv')
df.head()
```

```python
# Separting data and labels
x= df.drop(columns='Outcome', axis=1)
y= df['Outcome']
```

```python
print(x,y)
```

```python
# data standardization of X
scaler= StandardScaler()
scaler.fit(x)
stand_x=scaler.transform(x)
print(stand_x)
```

```python
# spliting the data to train and test
x_train, x_test, y_train, y_test=train_test_split(stand_x,y,test_size=0.2, random_state=2, stratify=y)
print(stand_x.shape, x_train.shape, x_test.shape)
```

```python
# Training the model
classifier= svm.SVC(kernel='linear')
classifier.fit(x_train, y_train)
```

```python
# Evaluating train model accuracy
x_train_pred= classifier.predict(x_train)
train_acc= accuracy_score(x_train_pred, y_train)
print(train_acc)
```

```python
# Evaluating test model accuracy
x_test_pred= classifier.predict(x_test)
test_acc= accuracy_score(x_test_pred, y_test)
print(test_acc)
```

```python
# Prediction Model

input_data= (4,110,92,0,0,37.6,0.191,30)
input_data_as_np= np.asarray(input_data)
input_data_reshape= input_data_as_np.reshape(1,-1) # predicting for one instance
std_data= scaler.transform(input_data_reshape)
prediction= classifier.predict(std_data)
if prediction[0]==0:
  print(prediction,"Non Diabetic")
else:
  print(prediction,"Diabetic")
```

```python
#Training Model
model=LogisticRegression()
model.fit(x_train, y_train)
```

```python
# Accuracy on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction, y_train)
print('Accuracy on training data: ', training_data_accuracy)
```

```python
# Accuracy on testing data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction, y_test)
print('Accuracy on testing data: ', test_data_accuracy)
```
